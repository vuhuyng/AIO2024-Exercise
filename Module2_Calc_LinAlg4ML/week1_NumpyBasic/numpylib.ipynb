Câu hỏi 10: `vectorize`

```python
import numpy as np

def maxx(x,y):
    if x >= y:
        return x
    else:
        return y

a = np.array([5 ,7 ,9 ,8 ,6 ,4 ,5])
b = np.array([6 ,3 ,4 ,8 ,9 ,7 ,1])
pair_max = np.vectorize(maxx, otypes =[float])
print(pair_max(a,b))
```

Bài tập 2: Xử lý ảnh

Các câu hỏi phần này giải quyết bài toán chuyển ảnh màu về ảnh xám. Ví dụ được mô 

tả như hình sau:



Ảnh sẽ được biểu diễn bởi tập các điểm ảnh (Pixel) có giá trị từ 0 đến 255. Ảnh màu sẽ được biểu
diễn thành ma trận có kích thước (Height, Width, Channel); trong đó số Channel là 3 tương ứng là giá
trị biểu diễn cho các kênh màu R-Red, G-Green, B-Blue. Ảnh xám sẽ được biểu diễn thành ma trận
có kích thước (Height, Width). Vì vậy, để chuyển ảnh màu thành ảnh xám, chúng ta sẽ tổng hợp từ 3
kênh màu RGB thành 1 giá trị duy nhất trong khoảng 0 đến 255. Có 3 phương pháp chính để tính giá
trị chuyển đổi:
• Lightness: Tính giá trị trung bình của giá trị lớn nhất và nhỏ nhất cho các kênh màu: (max(R,G,B)+ min(R,G,B))/2

• Average: Tính giá trị trung bình của 3 kênh màu: (R+G+B)/3

• Luminosity: Nhân hệ số tương ứng của 3 kênh màu như sau: 0.21*R + 0.72*G + 0.07*B

Câu hỏi 12: `Lightness` method

```python
# 1. Download image
!gdown 1i9dqan21DjQoG5Q_VEvm0LrVwAlXD0vB

# 2. Import necessary libraries
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt

# 3. Read the image
img = mpimg.imread('dog.jpeg')

# 4. Convert the image to grayscale using Lightness method
def rgb_to_gray_lightness(rgb):
    return 0.5 * (np.max(rgb, axis=2) + np.min(rgb, axis=2))

gray_img_01 = rgb_to_gray_lightness(img)

# 5. Display the first pixel value of the grayscale image
print(gray_img_01[0, 0])


plt.imshow(gray_img_01, cmap='gray')
plt.show()
```

Câu hỏi 13: `Average` method

```python
# 1. Download image
!gdown 1i9dqan21DjQoG5Q_VEvm0LrVwAlXD0vB

# 2. Import necessary libraries
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt

# 3. Read the image
img = mpimg.imread('dog.jpeg')

# 4. Convert the image to grayscale using Average method
def rgb_to_gray_average(rgb):
    return np.mean(rgb, axis=2)

gray_img_02 = rgb_to_gray_average(img)

# 5. Display the first pixel value of the grayscale image
print(gray_img_02[0, 0])

plt.imshow(gray_img_02, cmap='gray')
plt.show()
```

Câu hỏi 12: `Luminosity` method

```python
# 1. Download image
!gdown 1i9dqan21DjQoG5Q_VEvm0LrVwAlXD0vB

# 2. Import necessary libraries
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt

# 3. Read the image
img = mpimg.imread('dog.jpeg')

# 4. Convert the image to grayscale using Luminosity method
def rgb_to_gray_luminosity(rgb):
    return 0.21 * rgb[:, :, 0] + 0.72 * rgb[:, :, 1] + 0.07 * rgb[:, :, 2]

gray_img_03 = rgb_to_gray_luminosity(img)

# 5. Display the first pixel value of the grayscale image
print(gray_img_03[0, 0])

plt.imshow(gray_img_03, cmap='gray')
plt.show()
```

Câu hỏi 20: Gọi giá trị trung bình của cột Sales là A. Tạo ra mảng mới scores chứa các
giá trị Good, Average và Bad sao cho: nếu giá trị hiện tại > A => giá trị trong mảng mới
là Good, < A thì sẽ là Bad và = A sẽ là Average. Sau đó in ra kết quả scores[7:10]

```python
A = np.mean(df['Sales'])

scores = []
for sale in df['Sales']:
    if sale > A:
        scores.append('Good')
    elif sale < A:
        scores.append('Bad')
    else:
        scores.append('Average')

print(f"scores[7:10]: {scores[7:10]}")
```
