```python
data.filter(
    (data['guestroom'] == 'yes') &
    (data['basement'] == 'no')
)[['price', 'area', 'bedrooms', 'stories']]
```

## 8. Groupby operations

```python
data.group_by('price').agg(pl.col('area').mean()).head()
```

## 9. Sorting operations

```python
data.group_by('price').agg(pl.col('area')
    .mean()).sort(['area'], descending=True).head()
```

## 13. Applying functions

```python
def binary_2_numeric(decision):
    if decision == 'yes':
        return 1
    else:
        return 0

applied_data = data.with_columns(
    pl.col('mainroad').map_elements(binary_2_numeric),
    pl.col('guestroom').map_elements(binary_2_numeric),
    pl.col('basement').map_elements(binary_2_numeric),
    pl.col('hotwaterheating').map_elements(binary_2_numeric),
    pl.col('airconditioning').map_elements(binary_2_numeric),
    pl.col('prefarea').map_elements(binary_2_numeric)
)
applied_data
```
